```python
# !pip install openpyxl
import pandas as pd 
import datetime
```

```python
historical_cleaned = pd.read_parquet("./data_cleaned/yield_curve_historical_rates_MASTER.parquet")

historical_cleaned = (historical_cleaned
                      [historical_cleaned["Date"] <= "2024-03-28"]
                      .copy()
                      .reset_index()
                      .drop(columns=["index"])) 

historical_cleaned.head()
```

```python
df = pd.read_excel("./data_raw/daily-treasury-rates.xlsx")
df["Date"] = pd.to_datetime(df["Date"]).dt.strftime("%Y-%m-%d")
df = df[df["Date"] > "2024-03-28"].copy()
df["2 Yr - 10 Yr"] = df["2 Yr"] - df["10 Yr"] 
df["2 Yr - 30 Yr"] = df["2 Yr"] - df["30 Yr"] 
df["10 Yr - 30 Yr"] = df["10 Yr"] - df["30 Yr"] 
df.drop(columns = ["4 Mo"], inplace = True)
df
```

```python
pd.concat([df, historical_cleaned]).to_parquet("./data_cleaned/yield_curve_historical_rates_MASTER.parquet", index=False)
```
